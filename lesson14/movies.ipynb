```python
low = movie_df["year"].min()
high = movie_df["year"].max()
low, high
```

```python
start = round(low - 5, -1)
end = round(high + 5, -1)
start, end
```

```python
movie_decades = {}
current = start
while current < end:
    df = movie_df[movie_df["year"].between(current, current + 9)]
    key = str(current) + "s"
    movie_decades[key] = df
    current += 10

movie_decades["1940s"]
```

```python
low = tv_df["year"].min()
high = tv_df["year"].max()
low, high
```

```python
start = round(low - 5, -1)
end = round(high + 5, -1)
start, end
```

```python
tv_decades = {}
current = start
while current < end:
    df = tv_df[tv_df["year"].between(current, current + 9)]
    key = str(current) + "s"
    tv_decades[key] = df
    current += 10

tv_decades["1970s"]
```

```python
len(tv_decades["1950s"])
for decade in tv_decades.keys():
    print(decade, len(tv_decades[decade]))
```

```python
for decade in movie_decades.keys():
    print(decade, len(movie_decades[decade]))
```

```python
tv_decades["2000s"]
movie_decades["2000s"]

twothousands = pd.concat([tv_decades["2000s"], movie_decades["2000s"]])
twothousands.sort_values("imDbRating", ascending=False)
```
