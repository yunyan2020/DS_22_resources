```python
import pandas as pd
import requests
```

```python
api_key = ""
movie_url = f"https://imdb-api.com/en/API/Top250Movies/{api_key}"
tv_url = f"https://imdb-api.com/en/API/Top250TVs/{api_key}"
```

```python
movie_res = requests.get(movie_url)
movie_res
```

```python
movie_data = movie_res.json()
movie_data
```

```python
movie_data = movie_data["items"]
movie_data
```

```python
movie_df = pd.DataFrame(movie_data)
movie_df
movie_df.to_json("movies.json")
```

```python
tv_res = requests.get(tv_url)
tv_data = tv_res.json()
tv_data = tv_data["items"]
tv_df = pd.DataFrame(tv_data)
tv_df.to_json("tv.json")
```
